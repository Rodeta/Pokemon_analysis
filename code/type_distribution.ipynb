## Mean and standard deviation

```python
total_mean = data["Total"].mean()
total_std = data["Total"].std()
hp_mean = data["HP"].mean()
hp_std = data["HP"].std()
atk_mean = data["Attack"].mean()
atk_std = data["Attack"].std()
def_mean = data["Defense"].mean()
def_std = data["Defense"].std()
sp_atk_mean = data["Sp. Atk"].mean()
sp_atk_std = data["Sp. Atk"].std()
sp_def_mean = data["Sp. Def"].mean()
sp_def_std = data["Sp. Def"].std()
speed_mean = data["Speed"].mean()
speed_std = data["Speed"].std()

print(f"Total mean: {total_mean}, total std: {total_std}")
print(f"HP mean: {hp_mean}, HP std: {hp_std}")
print(f"Attack mean: {atk_mean}, attack std: {atk_std}")
print(f"Defense mean: {def_mean}, defense std: {def_std}")
print(f"Special attack mean: {sp_atk_mean}, special attack std: {sp_atk_std}")
print(f"Special defense mean: {sp_def_mean}, special defense std: {sp_def_std}")
print(f"Speed mean: {speed_mean}, speed std: {speed_std}")
```

## Legendary distribution

```python
labels = ["Legendary", "Not legendary"]
values = [data["Legendary"].value_counts()[1], data["Legendary"].value_counts()[0]]

fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels, autopct='%1.1f%%', startangle=360)

ax1.axis("equal")
plt.title("Distribution legendary pokemon")
plt.show()
```

## Types per generation bar chart

```python
data_gen_type = data[['Name', 'Type 1', 'Generation']].copy()
df_counts = data_gen_type.groupby(['Generation', 'Type 1']).count()
df_counts.rename(columns={'Name':'Count'}, inplace=True)
df_counts.reset_index(inplace=True)
```

```python
plt.figure(figsize=(14,8))
sns.barplot(x='Generation', y='Count', hue='Type 1', palette=type_colors, data=df_counts)
plt.title('Number of Pokemon of Each Type Added in Each Generation')
plt.ylabel('Number of Pokemon')
plt.show()
```

### Cumulative view

```python
df_counts.sort_values(['Type 1', 'Generation'], inplace=True)

# Calculate the cumulative counts
df_counts['Cumulative Count'] = df_counts.groupby('Type 1')['Count'].cumsum()

# Now you can create the plot, using 'Cumulative Count' instead of 'Count'
plt.figure(figsize=(14,8))
sns.barplot(x='Generation', y='Cumulative Count', hue='Type 1', palette=type_colors, data=df_counts)
plt.title('Cumulative Number of Pokemon of Each Type Added up to Each Generation')
plt.ylabel('Cumulative Number of Pokemon')
plt.show()
```

## Legendary types per generation bar chart

```python
data_legend = data[data['Legendary'] == True]
data_gen_type_legend = data_legend[['Name', 'Type 1', 'Generation', 'Legendary']].copy()
df_counts_legend = data_gen_type_legend.groupby(['Generation', 'Type 1']).count()
df_counts_legend.rename(columns={'Name':'Count'}, inplace=True)
df_counts_legend.reset_index(inplace=True)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Generation', y='Count', hue='Type 1', palette=type_colors, data=df_counts_legend)
plt.title('Number of legendary Pokemon of Each Type Added in Each Generation')
plt.ylabel('Number of Pokemon')
plt.show()
```

```python
# Create a pivot table
pivot_df = df_counts_legend.pivot_table(values='Count', index='Generation', columns='Type 1', aggfunc='sum').fillna(0)
# Make the data cumulative over generations
cumulative_df = pivot_df.cumsum()
colors = [type_colors[col] for col in cumulative_df.columns]
# Plot the data
cumulative_df.plot(kind='bar', stacked=True, figsize=(10, 8), color=colors)
plt.title('Cumulative count of legendary Pokemon by type and generation')
plt.xlabel('Generation')
plt.ylabel('Cumulative Count')
plt.legend(title='Type')
plt.xticks(rotation=0)
plt.show()
```

```python
pivot_df = df_counts_legend.pivot_table(values='Count', index='Generation', columns='Type 1', aggfunc='sum').fillna(0)
 
# Make the data cumulative over generations
cumulative_df = pivot_df.cumsum()
cumulative_df = cumulative_df.reset_index().melt('Generation', var_name='Type', value_name='Count')

plt.figure(figsize=(10, 8))
sns.barplot(data=cumulative_df, x='Generation', y='Count', hue='Type', palette=type_colors)

plt.title('Cumulative count of legendary Pokemon by type and generation')
plt.xlabel('Generation')
plt.ylabel('Cumulative Count')
plt.legend(title='Type')
plt.show()
```
